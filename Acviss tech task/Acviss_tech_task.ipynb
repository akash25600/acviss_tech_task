1. Function to Reverse a String

```python
def reverse_string(s):
    return s[::-1]

# Example usage:
print(reverse_string("Acviss technologies"))
```

2. Function to Find the Factorial of a Number

```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n-1)

# Example usage:
print(factorial(5))
print(factorial(10))
print(factorial(15))
```

3. Script to Sort a List of Integers in Ascending Order

```python
def sort_list(nums):
    return sorted(nums)

# Example usage:
print(sort_list([4, 2, 9, 1, 5, 6]))
print(sort_list([5, 2, 9, 1, 3, 6]))
print(sort_list([9, 2, 8, 1, 7, 6]))
```

6. Simple Binary Search Algorithm

```python
def binary_search(arr, target):
    low = 0
    high = len(arr) - 1

    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1

    return -1

# Example usage:
print(binary_search([1, 2, 3, 4, 5, 6, 7], 5))
print(binary_search([1, 2, 3, 4, 5, 6, 7], 6))
print(binary_search([1, 2, 3, 4, 5, 6, 7], 7))
```

8. Working with Databases: SQLite Example

```python
import sqlite3

def interact_with_db():
    conn = sqlite3.connect('example.db')
    c = conn.cursor()

    # Create table
    c.execute('''CREATE TABLE IF NOT EXISTS users (id INTEGER PRIMARY KEY, name TEXT)''')

    # Insert multiple values
    users = [('Alice',), ('Bob',), ('Charlie',), ('David',), ('Eve',)]
    c.executemany("INSERT INTO users (name) VALUES (?)", users)

    # Commit changes
    conn.commit()

    # Retrieve data
    c.execute('SELECT * FROM users')
    results = c.fetchall()

    c.execute('SELECT name FROM users')
    results = c.fetchall()

    conn.close()
    return results

# Example usage:
print(interact_with_db())

```

9. Python Script to Connect to SQLite, Insert Data, Retrieve It, and Perform a Simple Query

```python
import sqlite3

# Connect to SQLite database (or create it if it doesn't exist)
conn = sqlite3.connect('acviss.db')
cursor = conn.cursor()

# Create a table
cursor.execute('''
CREATE TABLE IF NOT EXISTS users (
    id INTEGER PRIMARY KEY,
    name TEXT,
    age INTEGER
)
''')

# Insert some data
cursor.execute('''
INSERT INTO users (name, age) VALUES (?, ?)
''', ('Alice', 30))

cursor.execute('''
INSERT INTO users (name, age) VALUES (?, ?)
''', ('Bob', 25))

# Commit the changes
conn.commit()

# Retrieve data
cursor.execute('SELECT * FROM users')
rows = cursor.fetchall()
print("Data from SQLite:")
for row in rows:
    print(row)

# Perform a simple query
cursor.execute('SELECT * FROM users WHERE age > ?', (28,))
rows = cursor.fetchall()
print("\nUsers older than 28:")
for row in rows:
    print(row)

# Close the connection
conn.close()
```

10. Script to Automate Renaming Multiple Files in a Directory Based on a Specific Pattern

```python
import os
import time
# Script to Automate Renaming Multiple Files in a Directory Based on a Specific Patter
def rename_files_with_timestamp(directory):
    # Get a list of all files in the directory
    files = os.listdir(directory)

    # Iterate through each file
    for file_name in files:
        # Get the full path of the file
        file_path = os.path.join(directory, file_name)

        # Check if the item is a file (not a directory)
        if os.path.isfile(file_path):
            # Get the current timestamp
            timestamp = int(time.time())

            # Create a new file name with the timestamp
            new_file_name = f"file_{timestamp}.txt"
            new_file_path = os.path.join(directory, new_file_name)

            # Rename the file
            os.rename(file_path, new_file_path)
            print(f"Renamed: {file_name} -> {new_file_name}")

# Example usage:
directory_path = '/content/sample_data'
rename_files_with_timestamp(directory_path)
```
